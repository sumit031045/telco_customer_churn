```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Telco_Customer_Churn_Dataset  (3).csv")
```

```python
df['Churn'].value_counts()
```

```python
Churn_rate = df['Churn'].value_counts(normalize=True) *100
Churn_rate
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Count")
plt.show()
```

```python
sns.countplot(x='gender',hue='Churn',data=df)
plt.title("Churn by Gender")
plt.show()
```

```python
sns.countplot(x='Partner',hue='Churn',data=df)
plt.title("Churn by Partner Status")
plt.show()
```

```python
sns.countplot(x='Dependents',hue='Churn',data=df)
plt.title("Churn by Dependents")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['tenure'],bins=30,kde=True)
plt.title("Tenure Distribution")
plt.show()
```

```python
sns.boxplot(x='Churn',y='tenure',data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
sns.countplot(x='Contract',hue='Churn',data=df)
plt.title("Churn by Contract Type")
plt.xticks(rotation=15)
plt.show()
```

```python
sns.countplot(x='PaymentMethod',hue='Churn',data=df)
plt.title("Churn by Payment Method ")
plt.xticks(rotation=45)
plt.show()
```

### Step 13: Observations & Insights

- Approximately 26.5% of customers have churned, while 73.5% are retained.
- Customers with month-to-month contracts show higher churn compared to long-term contracts.
- Customers with higher monthly charges tend to churn more frequently.
- These insights indicate that contract type and pricing play a key role in customer retention.
